```python
# To reduce the skewness of the above dataset
df1.info()
x=["Name",'Platform','Year','Publisher','Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales']
for i in x:
   df1[i]=np.log1p(df1[i])
```

```python
df1
```

```python
#removing outlayres using iqr method
import numpy as np
x=['NA_Sales','EU_Sales','JP_Sales','Other_Sales']
for i in x:
    data=df1[i]
    q1=np.quantile(data,0.25)
    q2=np.quantile(data,0.50)
    q3=np.quantile(data,0.75)
    iqr=q3-q1
    lower=q1-(1.5*iqr)
    upper=q3+(1.5*iqr)
    outliers=[j for j in data if j<lower or j>upper]
    print(len(outliers))
    df1= df1[(df1[i] >= lower) &(df1[i] <= upper)]
```

# Dimentionality Reduction

Reduce variables in a training dataset

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=['Name', 'Platform', 'Year', 'Publisher', 'NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales']
a=sc.fit_transform(df1[x])
pca=PCA(n_components=2)
pca_r=pca.fit_transform(a)
df1['pca']=pca_r[:,0]
df1["pca2"]=pca_r[:,1]
```

to remove the contents of the data using principal component analysis method

# Scaling

 process of transforming numerical features in a dataset to a common scale

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
df1[['Name', 'Platform', 'Year', 'Publisher', 'NA_Sales',
       'EU_Sales', 'JP_Sales', 'Other_Sales','pca','pca2']]=mm.fit_transform(df1[['Name', 'Platform', 'Year', 'Publisher', 'NA_Sales',
       'EU_Sales', 'JP_Sales', 'Other_Sales','pca','pca2']])
```

**process of transforming data values to a specfic range**

```python
# model evaluation
from sklearn.model_selection import train_test_split
y=df1["Global_Sales"]
x=df1.drop(columns=['Global_Sales'])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
print(f'x_train:{x_train.shape}\ny_train:{y_train.shape}\nx_test:{x_test.shape}\ny_test:{y_test.shape}')
```

```python
print(x_train)
y_train
# print(x_test)
y_test
```

# initial modeling and hypothesis

a hypothesis is a function that maps inputs to outputs

```python
from sklearn.svm import SVR
svm_model = SVR()
# fit the model to the train data
svm_model.fit(x_train,y_train)
# predict the target value of the test data
svm_prediction = svm_model.predict(x_test)
print("SVM_Prediction",svm_prediction)# y_test predict the value
```

predict the target value of the test dataset

# Documnetation and Reporting

```python
# calculating error
from sklearn.metrics import mean_absolute_error
svm_mae=mean_absolute_error(y_test,svm_prediction)
print("Mean absolute error of SVM = ",svm_mae)
```

Mean absolute error of support vector machine is 0.053791819941426976

```python
from sklearn.metrics import mean_squared_error
import math # Import the math module

svm_mse=mean_squared_error(y_test,svm_prediction)
svm_rmse=math.sqrt(svm_mse)  # Use math.sqrt to calculate the square root
print("Root Mean Squared Error of SVM",svm_rmse)
```

Root mean square error of the support  vector machine is  0.057175085706824696
